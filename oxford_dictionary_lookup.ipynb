```python
import pandas as pd
import numpy as np
pd.options.display.max_colwidth = 30000
pd.set_option('display.max_rows', 40)

# oxford api
import requests
import json
app_id = input("Enter Oxford dictionary Application ID: ")
app_key = input("Enter Oxford dictionary Application Key: ")
language = "en-gb"
```

```python
# load updated vocabulary file
vocabulary = pd.read_csv("Kindle_vocab_updated.csv")
```

```python
print(vocabulary.shape)
lookup = vocabulary.loc[0]['WORD']
print(lookup)
```

```python
def lookup_definition(word_num):
    # lookup word
    lookup = vocabulary.loc[word_num]['WORD']
    word_id = lookup
    url = "https://od-api.oxforddictionaries.com:443/api/v2/entries/" + language + "/" + word_id.lower() + "?fields=definitions&strictMatch=false"
    r = requests.get(url, headers={"app_id": app_id, "app_key": app_key})
    if r.status_code == 200:
        defin = r.json()
        try:
            definition = defin['results'][0]['lexicalEntries'][0]['entries'][0]['senses'][0]['definitions']
            add_def = vocabulary.iloc[word_num][1] = definition[0]
            print(definition[0])
        except Exception:
            print('************ Could not find definition.')
    
```

```python
# lookup definitions for vocabulary
num = 0
while num < (vocabulary.shape[0] - 1):
    if pd.isnull(vocabulary.iloc[num][1]):
        print(vocabulary.iloc[num][0])
        lookup_definition(num)
    num += 1
```
